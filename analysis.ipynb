```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],inplace=True,axis=1)
```

```python
covid_df.head()
```

```python
#Active Cases

covid_df['Active_Cases']=covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
covid_df.tail()
```

```python
#pivot table
#summing all deaths and cases for respective countries
statewise = pd.pivot_table(covid_df, values = ["Confirmed","Deaths","Cured"], index= "State/UnionTerritory", aggfunc="max") #aggfunc is aggregate function to sum all values
```

```python
#recovery rate or cured %
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
#ortality rate or death %
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
#sort in descending so since Confirmed is in the denominator , smallest recovery/death rate will be on the top
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
#matplotlib library has cmap == colormap for visualization
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
#TOP 10 Active cases variable

top_10_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values(by = ['Active_Cases'], ascending = False).reset_index()
```

```python
fig = plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 states with active cases in India", size="20")
```

```python
#seabornlibrary is for axes
#iloc is index location
ax=sns.barplot(data= top_10_active_cases.iloc[:10], y="Active_Cases",x="State/UnionTerritory", linewidth=2, edgecolor='red')
```

```python
top_10_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values(by = ['Active_Cases'], ascending = False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 states with most active cases in India", size=25)
ax=sns.barplot(data= top_10_active_cases.iloc[:10], y="Active_Cases",x="State/UnionTerritory", linewidth=2, edgecolor='red')

plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
# top 10 states with highest deaths

top_10_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by =['Deaths'],ascending=False).reset_index()

fig=plt.figure(figsize=(18,5))

plt.title("Top 10 states with most Deaths", size= 25)

ax = sns.barplot(data = top_10_deaths.iloc[:12], y = "Deaths", x="State/UnionTerritory", linewidth= 2, edgecolor = "red")
plt.xlabel("States")
plt.ylabel("Total Deaths")
plt.show()
```

```python
# Growth trend in the most no of confirmed cases

fig = plt.figure(figsize =(12,9))

ax= sns.lineplot(
    data= covid_df[
       covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka','Kerala', 'Tamil Nadu', 'Uttar Pradesh'])
    ], 
    x ='Date', y='Active_Cases', hue= 'State/UnionTerritory'
    )

ax.set_title("Top 5 States with Most Confirmed Cases", size=16)
```

```python
#clean data where state is india 
#state is given as india which is wrong 
vaccine = vaccine_df[vaccine_df.State!="India"]
vaccine
```

```python
vaccine.rename(columns = {"Total Individuals Vaccinated": "Total"}, inplace= True)
vaccine.head()
```

```python
# Most vaccinated State(top 5)
#QUESTION: When do you use ascending and descending
max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total',ascending=False)[:5] 
max_vac
```

```python
#top 10 states with highest deaths
fig=plt.figure(figsize=(10,5))

plt.title("Top 5 Vaccinated States", size= 20)

ax = sns.barplot(data = max_vac[:10], y = max_vac.Total, x=max_vac.index, linewidth= 2, edgecolor = "red")
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```

```python
min_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac = min_vac.sort_values('Total',ascending=True)[:5] 
min_vac
```

```python
#least 10 states with highest deaths
fig=plt.figure(figsize=(13,5))

plt.title("least 5 Vaccinated States", size= 20)

ax = sns.barplot(data = min_vac[:10], y = min_vac.Total, x=min_vac.index, linewidth= 2, edgecolor = "red")
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
